```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show, output_file
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.layouts import row

lemon = pd.read_csv('https://raw.githubusercontent.com/asukul/DS201/master/datasets/Lemonade2016-2.csv')
#Clean the Data
lemon.isnull().any() #are any of them null?
lemon.duplicated().any() #are there any duplicates?

#Add Sales & Revenue Column
lemon['Sales'] = lemon['Lemon'] + lemon['Orange']
lemon['Revenue'] = lemon['Sales'] * lemon['Price']
lemon['Profit'] = lemon['Revenue']/2
print(lemon.head())

#Make and show chart 1
source = ColumnDataSource(lemon)
output_notebook()
output_file("Taylor-DS201X-HW5-plot1.html")
q = figure(title = 'Sales by Temp, Revenue as Circle Size')
q.circle("Temperature", "Sales", source = source, fill_color = 'orange', size = "Profit", alpha = 0.5)
show(q)
```

```python
output_file("Taylor-DS201X-HW5-plot2.html")
q1 = figure(plot_width=250, plot_height=250)
q1.circle("Leaflets", "Revenue", source = source, fill_color = 'blue')

q2 = figure(plot_width=250, plot_height=250)
q2.circle("Sales", "Revenue", source = source, fill_color = 'green')

q3 = figure(plot_width=250, plot_height=250)
q3.circle("Temperature", "Sales", source = source, fill_color = 'red')

show(row(q1, q2, q3))
```
