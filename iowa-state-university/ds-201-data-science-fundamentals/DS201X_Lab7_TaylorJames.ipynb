```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read Data
df = pd.read_csv("https://raw.githubusercontent.com/asukul/DS201/master/datasets/Automobile_price_data_Cleaned.csv")

#Examine Data
df.head()
df.describe()
```

```python
plt.hist(df['city-mpg'], bins = 15)
plt.show()
```

```python
df1 = df['city-mpg']
df1.plot.hist(bins = 6)
plt.title('Automotive city-mpg')
plt.show()
```

```python
df.boxplot(column='city-mpg', by = 'make', rot = 45, figsize = (10,7), fontsize = 10, grid = False)
plt.show()
```

```python
df.plot.scatter(x = 'city-mpg', y = 'engine-size', c = 'curb-weight')
plt.show()
```
