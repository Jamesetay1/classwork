```python
import pandas as pd
import matplotlib.pyplot as plt

#Read The Data
df = pd.read_csv("http://bit.ly/drinksbycountry")
#print(df)

#What countries have beer consumption higher than 50?
df.loc[(df['continent'] == 'Europe') & (df['beer_servings'] > 50), ['country']]
print(df.loc[(df['continent'] == 'Europe') & (df['beer_servings'] > 50), ['country']])

#Group data by continent and create a default bar chart
df2 = df.groupby(['continent']).size()
df2.plot.bar()
plt.title("# of countries by continent")
plt.ylabel("# of countries")
plt.show()

#Use only "beer_servings" and group it by continent
#Create a bar graph of "beer servings" and group it by continent
df3 = df.groupby("continent").beer_servings.mean()
plt.ylabel("Mean beer servings")
plt.title("Mean beer servings by continent")
df3.plot.bar()
plt.show()

#Create a Box Plot of beer_servings by continent
df5 = df.boxplot(column="beer_servings", by = 'continent', rot = 45, figsize = (10,7), fontsize = 10, grid = False)
plt.ylabel("Beer servings")
plt.show()
```
