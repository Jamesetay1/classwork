```python
import pandas as pd
import numpy as np

#Import the data
lemon = pd.read_csv('Lemonade_Lab8.csv')

#Clean the Data
lemon.isnull().any() #are any of them null?
lemon.duplicated().any() #are there any duplicates?

#Bokeh Scatter Plot

#prepare data
from bokeh.models import ColumnDataSource
source_Q4 = ColumnDataSource(lemon)

#set output generation place
from bokeh.io import output_notebook, show
output_notebook()

#Use figure to generate plot
from bokeh.plotting import figure
p = figure(title = 'Lemon and Orange Sales by Temperature')

#add renderers
p.circle("Temperature", "Lemon", source = source_Q4, color = 'green', size = 8, legend = dict(value = "Lemon"))
p.triangle("Temperature", "Orange", source = source_Q4, color = 'orange', size = 8, legend = dict(value = "Orange"))
p.legend.location = "top_left"

#Output
show(p)
```

### QUESTION 5: 
###### It appears that there is some correlation between sales and temperature in general (not all that strong, but a loose trend).  Lemon also appears to outperform Orange by a small margin

```python

#std: avg-mean to avg+mean
# CREATE A VBAR CHART

from bokeh.models import ColumnDataSource, CDSView, GroupFilter
source_Q6 = ColumnDataSource(lemon)
park = CDSView(source = source_Q6, filters = [GroupFilter(column_name = 'Location', group = 'Park')])
beach = CDSView(source = source_Q6, filters = [GroupFilter(column_name = 'Location', group = 'Beach')])
p1 = figure(title = "Temperature vs Lemon Sales (Park and Beach)")
p1.triangle('Temperature', 'Lemon', source = source_Q6, color = 'blue', size = 8, view = park, legend = 'Sale in Park')
p1.circle('Temperature', 'Lemon', source = source_Q6, color = 'red', size = 8, view = beach, legend = 'Sale in Beach')

temperatures = list(lemon.groupby('Temperature').groups)
avg = lemon.groupby('Temperature')['Lemon'].mean()
std = lemon.groupby('Temperature')['Lemon'].std()
p1.vbar(x = temperatures, top = avg+std, width = 0.8, bottom = avg - std, fill_alpha = 0.2, legend = "Lemon Revenue 1 stddev")
p1.legend.location = 'top_left'

show(p1)


```

### QUESTION 7: 
###### Again, it appears that there is some correlation between sales and temperature in general (not all that strong, but a loose trend).  But also the park outpeforms the beach in terms of sales at lower temperatures (which makes sense because people are less likely to go to the beach when it is colder
