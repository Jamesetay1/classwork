```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('lab5.csv')
#print(df.head())
df.shape
df2 = df.groupby(['State_name']).size()
type(df2)
#print(df2)

df2.plot.bar(figsize =(15,7))
plt.title('# of Universities by State', fontsize = 20)
plt.xlabel('State Name', fontsize = 18)
plt.ylabel('# of Universities', fontsize = 18)
plt.show
```
