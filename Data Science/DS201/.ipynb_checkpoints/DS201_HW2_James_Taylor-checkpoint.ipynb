```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;

#Import the Data
sales = pd.read_csv("Lemonade2016-2.csv")

#Cleanse the Data
sales.iloc[8, 0] = "7/8/2016"
sales.iloc[20, 5] = round(sales.loc[:,"Leaflets"].mean())
sales = sales.drop_duplicates()

#Add Derived Columns
sales["Sales"] = sales["Lemon"] + sales["Orange"]
sales["Revenue"] = sales["Sales"] * sales["Price"]

#Create charts

#Scatter
plt.scatter(sales["Leaflets"], sales["Sales"], color = 'green')
plt.title("Leaflets vs Sales")
plt.xlabel("Leaflets")
plt.ylabel("Sales")
plt.show()

#Hitogram
plt.hist(sales["Revenue"], bins = 10, color = 'green')
plt.title("Revenue Histograms of Lemonade and Orange Sales")
plt.ylabel("Days")
plt.xlabel("$ in Revenue")
plt.show()

#Box and Whisker Plot based on Sales
plt.boxplot([sales.loc[sales['Location']=="Park", 'Sales'], sales.loc[sales['Location']=="Beach", 'Sales']])
plt.xticks([1,2],['Park', 'Beach'])
plt.xlabel("Location")
plt.title("Sales by location")
plt.ylabel("Sales")
plt.show()

#Box and Whisker Plot based on Sales and Temperature > 75
sales_above_75 = sales.loc[sales['Temperature'] >= 75, : ]
plt.boxplot([sales_above_75.loc[sales_above_75['Location']=="Park", 'Sales'], sales_above_75.loc[sales_above_75['Location']=="Beach", 'Sales']])
plt.xticks([1,2],['Park', 'Beach'])
plt.xlabel("Location")
plt.title("Sales by location when Temperature was >75F")
plt.ylabel("Sales")
plt.show()

#Seaborn Pair Plot
sns.set(style = 'ticks', color_codes = True)
sales_pairplot = sns.pairplot(sales, hue = 'Location', vars = sales[['Revenue', 'Temperature','Leaflets']], diag_kind = 'hist')
```
